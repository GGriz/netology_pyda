

6. Выдвинуть несколько гипотез по набору имеющихся переменных и проверить эти гипотезы с использованием t-test. 

6.1)  Атмосферные двигатели по трассе расходуют столько же сколько турбированные

```python
atmo_hw=df[df["aspiration"]=="std"]["highway-mpg"]
turbo_hw=df[df["aspiration"]=="turbo"]["highway-mpg"]
```

```python
len(atmo_hw), len(turbo_hw)
```

```python
atmo_hw = atmo_hw.sample(35)
turbo_hw = turbo_hw.sample(35)
```

```python
statistics.variance(atmo_hw), statistics.variance(turbo_hw)
```

```python
sts.ttest_ind(atmo_hw, turbo_hw, equal_var=False)
```

6.2)  Зависитли ли ширина машины ('width') от размера двигателя (engine-size).

```python
sts.ttest_ind(
    df['width'],
    df['engine-size'],
)
```

По полученным данным делаем вывод, что ширины машины не зависит от размера двигателя.
