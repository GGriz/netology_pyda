### Задание 1

Придумайте функции со следующими свойствами (в a) b) c) область определения X любая, какая вам удобна, главное - функция с нужными свойствами):

a) бесконечное количество локальных экстремумов

b) 2 локальных экстремума

с) 3 локальных экстремума

d*) область определения функции - 1 точка

###### решение

a) $f(x) = sin(x)$  
бесконечное количество локальных экстремумов

```python
def f(x):
    return np.sin(x)

x1 = np.linspace(-10.0, 10.0, 1000)
y1 = [f(x) for x in x1]
plt.plot(x1, y1)
```

b) $f(x) = 4sin(2x)+x^2$  
2 локальных экстремума на участке [-7, 7]

```python
def f(x):
    return 4 * np.sin(2*x) + x**2

x1 = np.linspace(-7.0, 7.0, 1000)
y1 = [f(x) for x in x1]
plt.plot(x1, y1)
```

c) $f(x) = 6sin(2x)+x^2$  
3 локальных экстремума на участке [-7, 7]

```python
def f(x):
    return 6 * np.sin(2*x) + x**2

x1 = np.linspace(-7.0, 7.0, 1000)
y1 = [f(x) for x in x1]
plt.plot(x1, y1)
```

d) $f(x) = 1$  
область определения функции - 1 точка 

```python
def f(x):
    return 1

x1 = np.linspace(-7.0, 7.0, 1000)
y1 = [f(x) for x in x1]
plt.plot(x1, y1)
```

### Задание 3:

Используя numpy, посчитайте значение 3-ей и 4-ой производных функций в точке:

$f(x) = x^5+4sin(2x)+cos(3x+3)$ в точке $x=1$


```python
from scipy.misc import derivative
from math import sin, cos
def f(x):
    return np.power(x,5) + 4*sin(2*x) + cos(3*x + 3)

print ('значение 3-ей производной функции в точке 1 равна {}'.format(derivative(f, 1.0, n=3, order=15)))
print ('значение 4-ой производной функции в точке 1 равна {}'.format(derivative(f, 1.0, n=4, order=15)))
```
